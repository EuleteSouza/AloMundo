# Treino e Teste

```python
# preparar os dados
# buscar valores considerando indices, ex: com : busca todas as linhas, da coluna 0
X = df.iloc[:,0].values
```

```python
print(f"anos de experiência: {X}")
```

```python
# para y

y = df.iloc[:,1].values
```

```python
print(f"salário: {y}")
```

Separando os dados (Treino e Teste)





```python
# importando sklearn
from sklearn.model_selection import train_test_split
```

### train_test_split: sepra treino e teste
(validação cruzada)

- dados a gente chama de X
 (Dados de treino - X_train // Dados de teste - X_test)
 iA supervisionada precisa de rótulos.
 ----------------------------------------------------------
- classes (rótulos) a gente chama de y



https://learn.microsoft.com/pt-br/training/modules/fundamentals-machine-learning/2-what-is-machine-learning

```python
# random so para ter um pont de partida pra aprendizagem
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=1/3,random_state=0)
```

```python
# importar o algoritmo
from sklearn.linear_model import LinearRegression
```

```python
# criar um objeto da classe linear regression para chamar os métodos (fit para treinar ) (predict para prever) entre outros
regressor = LinearRegression()
```

```python
# antes de treinar tem que transformar como matriz para nao dar erro
X_train = X_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)
```

```python
# treinando (primeiro treina)
regressor.fit(X_train,y_train)
```

```python
X_test = X_test.reshape(-1,1)
```

```python
# Na hora de prever,  chamar o rotulo criando uma variavel

y_pred = regressor.predict(X_test)
```

```python
# visualizar um gráfico
# finalizando com o grafico

import matplotlib.pyplot as plt

```

```python
# Visualizar os resultados de treino
plt.figure(figsize=(15,8))
plt.plot(X, y, 'Dr')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salário x Experiência ')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário')
plt.show()

```
